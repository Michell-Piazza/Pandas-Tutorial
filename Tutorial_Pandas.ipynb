# 3. Séries

Uma série do pandas é uma estrutura de dados unidimensional que pode armazenar dados como strings, inteiros, floats e outros objetos Python. O método pandas.Series() é usado para criar séries do pandas. Nas séries, os índices são armazenados na primeira coluna e os dados na segunda coluna.



> Se não especificarmos o argumento 'índice' em **pd.Series**, os índices serão números inteiros começando do 0

Exemplo:



```python
lista = ["Não", "Não", 3, 8]

serie = pd.Series(data=lista)

print(type(serie))
print(serie)
```

> Os índices podem ser adicionados através do argumento 'index' na **pd.Series**.





```python
lista = ["Não", "Não", 3, 8]
indices = ['Laranja','Maçã','Banana','Limão']

frutas = pd.Series(data = lista, index = indices)
frutas
```


> É possível obter o número de linhas e colunas, dimensões  ou número de elementos, utilizando o **Series.nome_do_atributo_desejado.**




```python
print("frutas.shape: ", frutas.shape) # Número de linhas e colunas, para uma Series, ele retorna apenas o número de elementos, já que uma série é unidimensional.
print("frutas.ndim: ", frutas.ndim) # Número de dimensões
print("frutas.size: ", frutas.size) # Número de elementos
```

> Obtendo os valores e indíces de uma série

```python
print("frutas.values: ", frutas.values) # Valores
print()
print("frutas.index: ", frutas.index) # Índices
```

> Verificando se um índice está disponível na série fornecida.

```python
print('Maçã' in frutas) # Existe o índice Maçã na série, logo retornou "True"
print('Abacaxi' in frutas) # Não existe o índice Abacaxi na série, logo retornou "False"
```

## Acessando, alterando e excluindo elementos na Série

Podemos acessar elementos em um DataFrame usando rótulos de índice ou índices numéricos dentro de colchetes [ ]. Índices negativos podem ser usados para acessar elementos do final da Série. O Pandas também disponibilizou os dois métodos, loc e iloc, para acessar elementos com índice de rótulo e índice numérico, respectivamente.

### Acesso usando rótulos de índice

```python
print(frutas)
print()

# Acessando os elementos da Frutas usando os indexs:

# Usando apenas um index
print("Usando index:\n", frutas['Maçã'])
print()

# Usando mais de um index
print("Usando indexs:\n", frutas[['Maçã', 'Laranja']])
print()

# O método .loc é usado para acessar um grupo de linhas e colunas por meio de rótulos
print("Usando .loc:\n", frutas.loc[['Banana', 'Laranja']])
print()

# Usando index numérico:
print("Usando index numérico:\n", frutas[[0, 1]])
print()

# Index negativo
print("Index negativo:\n", frutas[[-1]])
print()

# Index único
print("Index único:\n", frutas[0])
print()

#  O método .iloc é usado para acessar um grupo de linhas e colunas por meio de índices inteiros.
print("Usando .iloc:\n", frutas.iloc[[2, 3]])
```

### Alterando elementos da série

> A série em Pandas é mutável, podemos alterar seus elementos.



```python
print("Sem alterar:\n")
print(frutas)
print()

frutas['Maçã'] = 5

print("Após alterar:\n")
print(frutas)
```

### Excluindo elementos da série

> Para excluir utilizaremos o .drop()

```python
print("Antes da exclusão:\n")
print(frutas)
print()

frutas.drop('Maçã', inplace = True)

print("Após a Exclusão:\n")
print(frutas)
```

## Operações Aritméticas
Podemos realizar operações aritméticas entre Séries do Pandas e números individuais. Também podemos aplicar funções matemáticas do Numpy e realizar operações aritméticas em elementos selecionados.

```python
cores = pd.Series(data = [5, 12, 49], index = ["Verde", "Vermelho", "Azul"])
print(cores)
print()

print("Somando 2:\n", cores + 2)
print()

print("Subtraindo 2:\n", cores - 2)
print()

print("Multiplicando por 3:\n", cores * 2)
print()

print("Divisão por 2:\n", cores / 2)
```

> Também é possivel calcular exponencial, elevação e raíz quadrada.



```python
print(cores)
print()

print("Exponencial:\n", np.exp(cores))
print()

print("Elevação a 2 (quadrado):\n", np.power(cores, 2))
print()

print("Raíz quadrada:\n", np.sqrt(cores))
```

> Aplicando as operações aritméticas em algum item específico da série.

```python
print(cores)
print()

print("Somando 10 no Verde: ", cores['Verde'] + 10)
print()

print("Subtraindo 5 utilizando o iloc para a posição 2 (Azul): ", cores.iloc[2] - 5)
print()

print("Vermelho e Azul multiplicado por 2: ", cores[['Vermelho', 'Azul']] * 2)
print()

print("Divisão do Verde e Vermelho por 2 usando o loc:\n", cores.loc[['Verde', 'Vermelho']] / 2)
```

> É possível realizar operações aritméticas em Séries do Pandas com tipos de dados mistos, desde que a operação aritmética esteja definida para todos os tipos de dados na Série, caso contrário, receberemos um erro.

```python
lista = ["Não", "Sim", 20, 30]
indices = ['Vermelho', 'Verde', 'Azul', 'Preto']

cores = pd.Series(data = lista, index = indices)
print(cores)
print()

print("Multiplicando tudo por 2:\n", cores * 2)
```

# 4. DataFrames
O DataFrame do Pandas é uma estrutura de dados bidimensional com linhas e colunas rotuladas, que pode conter muitos tipos de dados. Um DataFrame é semelhante a uma planilha do Excel. Um DataFrame do Pandas pode ser criado manualmente ou carregado de um arquivo.

> O método pandas.DataFrame() é utilizado para criar um DataFrame manualmente. Primeiro, criaremos um dicionário e então o passaremos como um atributo para o pandas.DataFrame().





```python

dados = {
    "Maria": pd.Series(data=[55, 23, 67], index=['televisões', 'computadores', 'celulares']),
    "João": pd.Series(data=[17, 33, 42, 78], index=['livros', 'canetas', 'cadeiras', 'mesas'])
}

df = pd.DataFrame(dados)
print(type(df)) # Tipo do objeto (DataFrame)
print('----' * 10)  # Adicionando uma linha
print(df) # Mostrando o conteudo do DataFrame
```

>No exemplo acima, observamos que os rótulos das linhas do DataFrame 'df' estão ordenados alfabeticamente, e não na ordem que fornecemos no dicionário. Também podemos ver que existem alguns valores NaN. Mas o que são eles? NaN significa Not a Number e é a maneira do Pandas de indicar valores ausentes. Também podemos passar listas como chaves em vez de Séries do Pandas.


```python
dados = {"Manga" : [1, 2, 3, 20, 33],
       "Banana" : [3, 5, 8, 17, 21]}

dados_frutas = pd.DataFrame(dados)

dados_frutas
```

#5. Valores nulos (NaN)

> Contagem dos nulos

O Pandas oferece alguns métodos e atributos para contar os valores NaN. O método DataFrame.isnull() retornará True se um valor no DataFrame tiver um valor NaN, caso contrário, False. DataFrame.isnull().sum() retornará o número de NaN em cada coluna e DataFrame.isnull().sum().sum() retornará o número total de valores NaN nesse DataFrame.

```python

data = [
    {'computers': 25, 'jeans': 40, 'necklaces': 30, 't-shirts': 20, 'sandals':12, 'dresses':50},
    {'necklaces': 15, 'sunglasses': 60, 'computers': 20, 'jeans':8, 't-shirts': 5, 'sandals':10, 'dresses':15},
    {'computers': 30, 'jeans': 35, 'necklaces': 40, 'sunglasses': 3, 'sandals':15}
]

df = pd.DataFrame(data, index=["loja X", "loja Y", "loja Z"])
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
print("O total de NaN é: ", df.isnull().sum().sum())
```

>É possível contar o número de registros com .count()

```python
df.count()
```

```python
print("Total de registros não nulos é", df.count().sum())
```

>Eliminando valores nulos

Como contamos o número de nulos, agora podemos eliminá-los ou substituí-los.

Podemos remover valores NaN com a ajuda do método dropna(). No método dropna(), quando axis = 0, removerá linhas com valores NaN e quando axis = 1, removerá colunas com valores NaN. Utilizando o inplace = True salvará as mudanças no DataFrame original.

```python
# "Dropando" as linhas com valores nulos (NaN)
df.dropna(axis = 0, inplace = True)
df
```

>Contagem de variavéis categóricas


Podemos contar as variáveis categóricas em um conjunto de dados. Isso pode ser feito com o método value_counts(). Também podemos obter a proporção de cada categoria no conjunto de dados passando o argumento normalize=True. Além disso, podemos classificar os resultados passando o argumento sort=True.

```python
#Contagem de registros em cada estados
state_counts = df_limpo['STATE'].value_counts(sort = True)
state_counts
```

O parâmetro normalize no método .value_counts() do Pandas é usado para normalizar os valores contados, ou seja, ele converte as contagens em proporções relativas. Quando normalize=True é especificado, o método irá retornar a frequência de cada valor como uma proporção do total de observações. Isso é útil para obter uma visão rápida da distribuição dos valores em uma determinada coluna, expressa como uma porcentagem do total.

```python
#Contagem de registros em cada estado, normalizados
state_counts = df_limpo['STATE'].value_counts(normalize = True, sort = True)
state_counts
```

>Agrupamento


O método .groupby() do Pandas é usado para agrupar os dados com base em uma ou mais variáveis. Ele divide o DataFrame em grupos com base nos valores exclusivos de uma ou mais colunas e, em seguida, é possível aplicar uma função a esses grupos.

```python
# Agrupando os dados por estado e calculando o IDHM médio
estado_agrupado = df.groupby("STATE")["IDHM"].mean()

# Ordenando o resultado em ordem decrescente
estado_agrupado_ordenado = estado_agrupado.sort_values(ascending=False)

# Exibindo o resultado
print(estado_agrupado_ordenado)
```

```python
# Agrupando os dados por estado e encontrando o estado com maior número de pessoas com 60+ de idade segundo o IBGE
estado_agrupado = df.groupby("STATE")["IBGE_60+"].max()

# Ordenando o resultado em ordem decrescente
estado_agrupado_ordenado = estado_agrupado.sort_values(ascending=False)

# Exibindo o resultado
print(estado_agrupado_ordenado)
```

>Estatísticas Múltiplas agrupadas


Também podemos calcular várias estatísticas resumidas agrupadas usando o método agg(). O método agg() aplica uma função ou uma lista de funções a uma Série ou uma coluna de um DataFrame.

```python
# Agrupando os dados por estado e calculando várias estatísticas resumidas para a coluna "IDHM"
grouped_category = df.groupby("STATE")["IDHM"].agg([np.mean, np.median, np.min, np.max])

# Exibindo o resultado
grouped_category
```
